```python
#Glass type prediction using Random Forest algorithm
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/dataset3/main/glass.csv',header=None)
df
```

```python
df.columns =['Index','RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type']
df
```

```python
Y = df["Type"]
Y.head()
df = df.drop("Type",axis=1)
df.head()
```

```python
from sklearn import preprocessing  
t = df.values
min_max_scaler = preprocessing.MinMaxScaler()
t_scaled = min_max_scaler.fit_transform(t)
df = pd.DataFrame(t_scaled)
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test= train_test_split(df,Y,test_size = 0.40, random_state= 0)
print((X_train.shape))
```

```python
import math 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from math import sqrt
from sklearn import preprocessing
from sklearn import utils
from sklearn.metrics import accuracy_score,classification_report
#convert y values to categorical values
clf= RandomForestClassifier(n_estimators=500,max_depth=4,random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
asc=accuracy_score(y_test, y_pred)
print(classification_report(y_test,y_pred))
```
